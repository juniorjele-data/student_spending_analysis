## Exploratory Data Analysis

This section explores overall spending patterns across categories.

```python
df.sum(numeric_only=True).plot(kind='bar')
plt.title("Clean Spending by Category")
plt.show()
```

### Insight

Tuition is the largest expense among students, significantly higher than other categories.

## Group Analysis

This section compares spending across different groups such as year and major.


```python
import matplotlib.pyplot as plt

df.groupby('year_in_school')['food'].mean().plot(kind='bar')
plt.title("Food Spending by Year")
plt.xlabel("Year")
plt.ylabel("Average Food Spending")
plt.show()

```

### Food Spending by Year

Seniors tend to spend the most on food, while juniors spend the least. Freshmen and sophomores fall in between.



```python
df.groupby('major')[['technology','food']].mean()
```

### Food Spending by Year

Seniors tend to spend the most on food, while juniors spend the least. Freshmen and sophomores fall in between.

### Technology Spending by Major

Biology students spend the most on technology, while Computer Science students spend the least. This may be because CS students already own necessary devices.


```python
df.groupby('major')['technology'].mean()
```

## Correlation Analysis

There is little to no correlation between income and spending categories such as food and entertainment.





## Prediction Model

A linear regression model was used to predict food spending based on monthly income.

```python
import matplotlib.pyplot as plt

plt.scatter(df['monthly_income'], df['food'])
plt.plot(df['monthly_income'], model.predict(X), color='red')
plt.xlabel("Monthly Income")
plt.ylabel("Food Spending")
plt.title("Income vs Food Spending")
plt.show()
```

### Prediction Insight

The model shows that income is not a strong predictor of food spending, as indicated by the nearly flat regression line.

## Conclusion

This project demonstrates that tuition is the largest expense for students. Additionally, income does not significantly influence spending behavior. Differences in spending were observed across academic years and majors. The predictive model further confirmed the weak relationship between income and food spending.
